#### Q1: Basic Filtering

````python
employees[(employees['salary'] > 80000) & (employees['department'] == 'Engineering')]
````

#### Q2: isin + Rating Filter

````python
employees[(employees['department'].isin(['HR', 'Sales'])) & (employees['rating'] >= 4.0)]
````

#### Q3: String Filter

````python
employees[employees['name'].str[0].str.upper().isin(list('AEIOU'))]
````

#### Q4: Top N per Group

````python
employees.groupby('department').apply(
    lambda x: x.nlargest(2, 'salary'), include_groups=False
).reset_index(level=0)
````

#### Q5: query() Method

````python
employees.query('hire_date > "2020-01-01" and age < 30')
````

#### Q6: Filter by Another DataFrame

````python
employees[employees['emp_id'].isin(sales['emp_id'].unique())]
````

#### Q7: Conditional Column

````python
df = employees.copy()
df['level'] = np.where(df['salary'] >= 80000, 'Senior', 'Junior')
df[['name', 'salary', 'level']]
````

#### Q8: between()

````python
sales[sales['amount'].between(1000, 1500)]
````

#### Q9: Basic GroupBy

````python
employees.groupby('department').agg(
    avg_salary=('salary', 'mean'),
    total_bonus=('bonus', 'sum')
).reset_index()
````

#### Q10: Multiple Aggregations

````python
employees.groupby('department')['salary'].agg(['min', 'max', 'mean', 'std']).reset_index()
````

#### Q11: Transform

````python
df = employees.copy()
dept_total = df.groupby('department')['salary'].transform('sum')
df['salary_pct_of_dept'] = (df['salary'] / dept_total * 100).round(2)
df[['name', 'department', 'salary', 'salary_pct_of_dept']]
````

#### Q12: GroupBy Filter

````python
employees.groupby('department').filter(lambda x: x['salary'].mean() > 75000)
````

#### Q13: Cumulative Sum

````python
df = sales.sort_values('sale_date').copy()
df['cum_amount'] = df.groupby('emp_id')['amount'].cumsum()
df[['emp_id', 'sale_date', 'amount', 'cum_amount']]
````

#### Q14: Rank Within Group

````python
df = employees.copy()
df['salary_rank'] = df.groupby('department')['salary'].rank(ascending=False, method='dense').astype(int)
df[['name', 'department', 'salary', 'salary_rank']].sort_values(['department', 'salary_rank'])
````

#### Q15: Percentage of Group Total

````python
df = sales.copy()
df['pct_of_product_total'] = (
    df['amount'] / df.groupby('product')['amount'].transform('sum') * 100
).round(2)
df[['sale_id', 'product', 'amount', 'pct_of_product_total']]
````

#### Q16: First & Last per Group

````python
sales.sort_values('sale_date').groupby('emp_id').agg(
    first_sale=('sale_date', 'first'),
    last_sale=('sale_date', 'last'),
    total_sales=('sale_id', 'count')
).reset_index()
````

#### Q17: Rolling Average

````python
df = stock_prices.sort_values(['ticker', 'date']).copy()
df['rolling_3d_avg'] = df.groupby('ticker')['close'].transform(
    lambda x: x.rolling(3, min_periods=1).mean()
)
df[['ticker', 'date', 'close', 'rolling_3d_avg']].head(10)
````

#### Q18: Lag & Lead

````python
df = sales.sort_values(['emp_id', 'sale_date']).copy()
df['prev_amount'] = df.groupby('emp_id')['amount'].shift(1)
df['next_amount'] = df.groupby('emp_id')['amount'].shift(-1)
df[['emp_id', 'sale_date', 'amount', 'prev_amount', 'next_amount']]
````

#### Q19: Percent Change

````python
df = stock_prices.sort_values(['ticker', 'date']).copy()
df['daily_return'] = df.groupby('ticker')['close'].pct_change().round(4)
df[['ticker', 'date', 'close', 'daily_return']].head(10)
````

#### Q20: Z-Score

````python
df = sales.copy()
df['mean_amount'] = df.groupby('emp_id')['amount'].transform('mean')
df['std_amount']  = df.groupby('emp_id')['amount'].transform('std')
df['z_score'] = ((df['amount'] - df['mean_amount']) / df['std_amount']).round(3)
df[['emp_id', 'sale_date', 'amount', 'mean_amount', 'z_score']]
````

#### Q21: Row Number Within Group

````python
df = sales.sort_values(['emp_id', 'sale_date']).copy()
df['sale_seq'] = df.groupby('emp_id').cumcount() + 1
df[['emp_id', 'sale_date', 'amount', 'sale_seq']]
````

#### Q22: Expanding Mean

````python
df = sales.sort_values(['emp_id', 'sale_date']).copy()
df['expanding_mean'] = df.groupby('emp_id')['amount'].transform(
    lambda x: x.expanding().mean()
)
df[['emp_id', 'sale_date', 'amount', 'expanding_mean']]
````

#### Q27: Quantile Bins

````python
df = employees.copy()
df['salary_quartile'] = pd.qcut(df['salary'], q=4, labels=['Q1','Q2','Q3','Q4'])
df[['name', 'salary', 'salary_quartile']]
````

#### Q28: Weighted Average

````python
df = sales.copy()
df['unit_price'] = df['amount'] / df['quantity']
df.groupby('product').apply(
    lambda x: np.average(x['unit_price'], weights=x['quantity']), include_groups=False
).reset_index(name='weighted_avg_price')
````

#### Q29: Left Join

````python
total_sales = sales.groupby('emp_id')['amount'].sum().reset_index(name='total_sales')
employees.merge(total_sales, on='emp_id', how='left').fillna({'total_sales': 0})
````

#### Q30: Anti Join

````python
employees[~employees['emp_id'].isin(sales['emp_id'].unique())]
````

#### Q31: Self Join

````python
merged = employees.merge(employees, on='department', suffixes=('_a', '_b'))
merged = merged[merged['emp_id_a'] < merged['emp_id_b']]
merged = merged[merged['salary_a'] >= merged['salary_b'] * 1.2]
merged[['name_a', 'name_b', 'department', 'salary_a', 'salary_b']]
````

#### Q32: Consecutive Condition

````python
df = sales.sort_values(['emp_id', 'sale_date']).copy()
df['above_1200'] = (df['amount'] > 1200).astype(int)
df['streak'] = df.groupby('emp_id')['above_1200'].transform(
    lambda x: x.groupby((x != x.shift()).cumsum()).cumsum()
)
df['hot_streak'] = df['streak'] >= 3
df[['emp_id', 'sale_date', 'amount', 'above_1200', 'streak', 'hot_streak']]
````

#### Q33: Month-over-Month Growth

````python
df = sales.copy()
df['month'] = df['sale_date'].dt.to_period('M')
monthly = df.groupby(['emp_id', 'month'])['amount'].sum().reset_index()
monthly['prev_month'] = monthly.groupby('emp_id')['amount'].shift(1)
monthly['mom_growth_pct'] = (
    (monthly['amount'] - monthly['prev_month']) / monthly['prev_month'] * 100
).round(2)
monthly
````
